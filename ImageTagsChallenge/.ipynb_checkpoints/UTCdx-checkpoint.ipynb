## *Question 1:* 

Suppose we randomly select a row from the tags.csv file. Write a function that takes as input any tag and returns the probability that it appears in the randomly selected row. Report the probability that the tag `thing` will appear in the randomly selected row.

For this I calculated the expected value of the probability using two methods to compare: straight from the dataset, and an MCMC method.

```python
def probability(data, word):
    """
    Returns the expected value of the probability from
    the supplied data set.
    
    Input: data (pandas dataframe), word (string) the
    word to calculate the probability of.
    
    Output: probability of word occurring in data.
    """
    
    if word not in data.columns:
        return 0.
    
    return data[word].mean()
    

def MCMC_probability(data, word, num_samples=100000):
    """
    Returns the expected value of the probability and 
    uncertainty of probability as standardized error 
    for "word" to occur in the dataframe data. I have
    assumed here that the central limit theorem holds
    and that each measurement is independent.
    
    Input: data (pandas dataframe), word (string) the
    word to calculate the probability of, and
    num_samples the number of times to sample the 
    data.
    
    Output: probability of word occurring in data
    along with the 1 sigma uncertainty.
    """
    
    if word not in data.columns:
        return 0. ,0.
    
    dice_throws = data.sample(num_samples, replace=True)
    
    mu_probability    = dice_throws[word].mean()
    delta_probability = dice_throws[word].sem()
    
    return mu_probability, delta_probability
```

Using these functions, the sample probability of querying a line with the word `thing` in it is:

```python
probability(data, 'thing')
```

While the MCMC method gives the expected result of an oscillating mean about the population average (shown below with $1\sigma$ uncertainty).

```python
means  = []
sigmas = []
x_range = range(1000, 10000, 500)

for k in tqdm(x_range):
    mu, delta = MCMC_probability(data, 'thing')
    means.append(mu), sigmas.append(delta)
    
sample_mean_probability = probability(data, 'thing')

fig, ax = plt.subplots()

ax.errorbar(x=x_range, y=means, yerr=sigmas, lw=0.8, ecolor='darkred', label=r'MCMC $<p> \pm \delta$')
ax.plot([x_range[0], x_range[-1]], [sample_mean_probability, sample_mean_probability], ls='--', lw=1.0, color='k', label=r'Sample $<p>$')

ax.legend(frameon=False)
fig.tight_layout()
```

## *Question 2:* 

Write a function that takes as input any two tags and returns the probability of finding the second tag in a row containing the first tag. Report the following two numbers:

•	(a) The probability of finding a tag named physical_entity in a row that contains the tag named object.

•	(b) The probability of finding a tag named `object` in a row that contains the tag named `physical_entity`.


*Answer 2:*

By conditioning p(word2) on p(word1), we arrive at our answer:

```python
def probability_together(data, word1, word2):

    if word1==word2:
        return 1.
    try:
        return probability(data[data[word1]==1], word2)
    except:
        return 0.
```

```python
probability_together(data, 'object', 'physical_entity')
```

In lines that contain the word `object`, there is a 1.0 probability that the word `physical object` will be found.

```python
probability_together(data, 'physical_entity','object')
```

Conditioned on lines containing `physical_entity`, there is a 0.22 probability the word `object` will be found.

## *Question 3:* 

Suppose we randomly select N rows from the tags.csv with replacement. Write a function that takes as input two positive integers K and N and returns the pairs of tags that have a good chance (greater than 50%) of appearing together on at least K rows from the random sample of N rows. Report any pair of tags that have a good chance of appearing together on at least 10 rows from a random sample of 100 rows.

*Answer:*

The problem may be solved in a straightforward way with simple applications of Bayes' Theorem and conditional probabilities. Below, I first take the conditional probability that word1 occurs in the data set, and multiply the sum density of word2 given word1 * the sample size $N$. 

```python
def highest_probability_pairs(data, N, K):

    sample_size = data.shape[0]
    pairs   = []

    for column in tqdm(data.columns):
        
        exp_value_sum      = data[data[column]==1.].loc[:,column:].sum() / sample_size * N
        high_chance_pairs  = exp_value_sum[(exp_value_sum > K) & (~exp_value_sum.index.isin([column]))].index.tolist()
        
        [pairs.append((column, x, exp_value_sum[x])) for x in high_chance_pairs]
        
    return pairs
```

```python
%%time
pairs = highest_probability_pairs(data, N=100, K=10)
```

The following pairs have a high chance of appearing K=10 times together with a subsample of N=100, reported along with the expectation values of their distribution sum:

```python
pairs
```

### Validating the algorithm

With the below code, we can validate the correctness of the algorithm by choosing an example pair. If you would like to test others, please adjust the pair definitions.

```python
word1, word2 = 'physical_entity', 'causal_agency'
num_MCMC_samples = 1000

avg_sum = []
for _ in range(num_MCMC_samples):
    
    sample = data.sample(100)
    avg_sum.append(sample[(sample[word1]==1) & (sample[word2]==1)][word2].sum())

print("\nThe pair ('{}', '{}') will occur (within 99.7% confidence):".format(word1, word2))
print('\n<S> = {} +- {}.'.format(np.mean(avg_sum), 3*np.std(avg_sum)/np.sqrt(num_MCMC_samples)))
```
